## K-Means clustering (Number of clusters(n) = 2)

```python
import sklearn
kmeans = sklearn.cluster.KMeans(n_clusters = 2, random_state = 12)
X = df[['Weight (kg)','Height (cm)']]
print(X)
```

```python
kmeans.fit(X)
df['Clusters'] = kmeans.predict(X)
centers = kmeans.cluster_centers_
print(centre)
```

```python
print(df)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X["Weight (kg)"], X["Height (cm)"], c=df['Clusters'], cmap='rainbow', alpha=0.6, s=100, label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='X', s=200, label='Cluster Centers')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('K-Means Clustering of Students')
plt.legend()
plt.grid(True)
plt.show()
```

### Extracting countries and indicator IDs for required data fetching

```python
import wbdata
import pandas as pd

sources = wbdata.get_sources()
countries = pd.DataFrame(wbdata.get_countries())
indicators = pd.DataFrame(wbdata.get_indicators())
with pd.ExcelWriter("world_bank_data.xlsx", engine='openpyxl') as writer:
    countries.to_excel(writer, sheet_name="Countries", index=False)
    indicators.to_excel(writer, sheet_name="Indicators", index=False)

print("Data saved to world_bank_data.xlsx")
```

### Defining Indicators, specifying time (1st January 2016 - 31st Decemeber 2020)
### Downloading dataset

```python
import datetime
start_date = datetime.datetime(2016, 1, 1)
end_date = datetime.datetime(2020, 12, 31)

# Define the indicators
indicators = {
    'NY.GDP.PCAP.CD': 'GDP per Capita',
    'SI.POV.GINI': 'Gini Coefficient',
    'SI.POV.DDAY': 'Poverty Headcount Ratio at $1.90/day',
    'SL.UEM.TOTL.ZS': 'Unemployment Rate',
    'GC.DOD.TOTL.GD.ZS': 'Debt-to-GDP Ratio'
}

# Define the countries of interest (ISO 3-letter codes)
countries = ['USA', 'CHN', 'DEU', 'JPN', 'IND', 'BRA', 'RUS', 'GBR', 'CAN', 'FRA']

# Fetch the data for the last 5 years
data = wbdata.get_dataframe(indicators, country=countries, date=(start_date, end_date))

# Reset index to make 'country' and 'date' as columns
data.reset_index(inplace=True)
print(data.head(n = len(data)))
```

```python
data.shape
data.isnull().sum()
```

```python
# Drop rows with missing values
data_clean = data.dropna()
data_clean.shape
```

### Data scaling

```python
from sklearn.preprocessing import StandardScaler

features = ['GDP per Capita', 'Gini Coefficient', 'Poverty Headcount Ratio at $1.90/day', 'Unemployment Rate', 'Debt-to-GDP Ratio']

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_clean[features])

print(pd.DataFrame(data_scaled, columns=features).head())
```

### Using Elbow Method to determine K cluster

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow Method to find optimal number of clusters
inertia = []  # Within-cluster sum of squares
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.plot(range(1, 11), inertia,marker = 'o', linestyle = '--')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-cluster Sum of Squares)')
plt.show()
```

### Wihtout using PCA

```python
kmeans = KMeans(n_clusters= 5, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Add the cluster labels to the original data
data_clean['Cluster'] = clusters

# Display the data with cluster labels
#print(data_clean.head())
```
